# Cleans 3 CSV files, connecting the readers's age by the genre
Prior to being added to this notebook, they have already been cleaned a bit by deleting unused columns

```python
# Retrieve all the datasets as a dataframe
amazon_df = pd.read_csv("../old/BookRec/Amazon_TitleToGenre.csv")
booksToTitle_df = pd.read_excel("../old/BookRec/BookRec_BooksToTitle.xlsx")
userToBook_df = pd.read_excel("../old/BookRec/BookRec_UserToBook.xlsx")
userAge_df = pd.read_csv("../old/BookRec/Cleaned-BookRec_UserAgeToCountry.csv")
```

```python
display(amazon_df)
# display(booksToTitle_df)
# display(userToBook_df)
# display(userAge_df)
```

```python
# Next start combining tables together, starting from the age table
merge_UserAge_UserToBook = pd.merge(userAge_df, userToBook_df, on="User-ID", how="inner")
display(merge_UserAge_UserToBook)
```

```python
# Next find the book title base on the ISBN numeber
merge_M1_BooksToTitle = pd.merge(merge_UserAge_UserToBook, booksToTitle_df, on="ISBN", how="inner")
display(merge_M1_BooksToTitle)
```

```python
# Now use the book title to find the genre for each reader
# Rename the column since they do not match
merge_M1_BooksToTitle = merge_M1_BooksToTitle.rename(columns={'Book-Title' : 'Title'})
merge_M2_Amazon = pd.merge(merge_M1_BooksToTitle, amazon_df, on="Title", how="inner")
display(merge_M2_Amazon)
```

```python
# Drop unused columns
cleaned_df = merge_M2_Amazon.drop(columns=['Unnamed: 0_x', 'Unnamed: 0_y'])
cleaned_df = cleaned_df.dropna()
display(cleaned_df)
```

```python
# Since there seems to be a duplicate for each record, delete the duplicate
cleaned2_df = cleaned_df.drop_duplicates(subset=['User-ID', 'Title'])
display(cleaned2_df)
```

```python
# Since the dataset of 1600 records is a bit too small, will need to check the age distributions
cleaned2_df["Age"].hist()
```

```python
# Add Age labels to divide the data
# Ranges: 0-18 (Teens), 18-30 (Young Adult), 30-60 (Adult), 60-120 (Seniors) 
age_bins = [12, 18, 30, 60, 120]

age_groups = ["Teens (12-17)", "Young Adult (18-29)", "Adult (30-59)", "Senior (60+)"]

cleaned3_df = cleaned2_df.copy()

cleaned3_df["Age_Group"] = pd.cut(cleaned2_df["Age"], bins=age_bins, labels=age_groups, right=False)

display(cleaned3_df)
```

```python
# Save the cleaned dataset
cleaned3_df.to_csv("../Combined_BookRec_GenreToAge.csv")
```
