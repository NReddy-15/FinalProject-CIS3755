 # Cleaning GlobalReading GenreToUser
 For users that have more than 1 favorite genre, create another record to represent that user in that genre 

The next step, S2, will clean and merge the user's gender and country to their favorite genres

```python
import pandas as pd
```

```python
#  Generate the df that has the GenreByUser
genreToUser_df = pd.read_excel('../old/GlobalReading_GenreToUser.xlsx')
```

```python
# Display the dataframes

# display Book_ID, Shelve_Names (Messy Genres list), User_ID
display(genreToUser_df)
```

```python
# View all the genres available
unique_genres = set()

# Loop through all the records' genre
for record in genreToUser_df['Shelve_Names']:
    if isinstance(record, str):
        # There could be more than one genre per user
        # The list is of type string, so we split by , 
        # to get each genre
        words = record.split(',')
        # Add the genre to the set, seeing what are unique
        unique_genres.update(words)

# Remove these weak genres
unique_genres.discard('to-read')
unique_genres.discard('SMALL_COUNT')
    
display(unique_genres)
```

# Create genre lists
Base on the given genres, I manually cleaned and group each keyword by a general genre group

```python
genre_groups = {
    "Fantasy": ['fantasy', 'urban-fantasy', 'mythology', 'vampires', 'paranormal'],
    "Sci-Fi": ['sci-fi', 'science-fiction', 'dystopia', 'dystopian'],
    "Mystery/Thriller": ['mystery', 'crime', 'thriller', 'suspense', 'true-crime'],
    "Romance": ['romance', 'chick-lit'],
    "Children": ['children', 'childrens', 'children-s-books', 'kids', 'picture-books'],
    "Young Adult": ['young-adult', 'ya'],
    "Non-Fiction": ['non-fiction', 'nonfiction', 'biography', 'memoir', 'history',
                    'historical', 'politics', 'economics', 'philosophy', 'psychology', 
                    'science', 'business', 'education', 'self-help', 'health', 
                    'parenting', 'travel', 'cookbooks', 'food'],
    "Historical Fiction": ['historical-fiction'],
    "Graphic Novels/Manga": ['comics', 'graphic-novel', 'graphic-novels', 'manga'],
    "Art/Literature": ['art', 'music', 'poetry', 'plays', 'drama', 'classic', 'classics', 'literature'],
    "Humor": ['humor', 'comedy'],
    "Short Stories": ['short-stories'],
    "Horror": ['horror'],
    "Adventure": ['adventure'],
}
```

```python

# Flatten the general genres, so now a dictionary is created 
# with the keywords being the key and the general genre being the value
flat_map = {word: category for category, words in genre_groups.items() for word in words}
display(flat_map)
```

```python
# Loop through and reorganize the shelve_names => Genres

def GetGeneralGenres(keywords):
    # Each user can have multiple favorite genres
    general_genres = []
    
    
    if  isinstance(keywords, str):
        # Get the current string of genres and split them into individual strings
        keywordsList = keywords.split(',')
        
        # Check what value the genre could be and convert it to the general general
        for word in keywordsList:
            if word in flat_map:
                genre = flat_map[word] 
                general_genres.append(genre)
                
        # Return a list of general genres
        return list(general_genres)
    return []
```

```python
# Create a copy of the genreToUser and apply the shelve name filter function above
cleaned_genre_df = genreToUser_df.copy()
cleaned_genre_df['Genres'] = cleaned_genre_df['Shelve_Names'].apply(GetGeneralGenres)
```

```python
# Displays Book_ID, Shelve_Names, User_ID, Genres
display(cleaned_genre_df)
```

```python
# Delete any records that have an empy list in genre
cleaned_genre_df = cleaned_genre_df[cleaned_genre_df['Genres'].str.len() > 0]
```

```python
# Displays Book_ID, Shelve_Names, User_ID, Genres
display(cleaned_genre_df)
```

```python
# Print and salve the new data
cleaned_genre_df.to_csv('../Cleaned_GlobalReader_GenreToUser.csv')
```
