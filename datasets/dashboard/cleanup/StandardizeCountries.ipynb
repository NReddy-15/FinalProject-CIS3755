```python
print(map_df["most_popular_country"])


def GetGeneralCountries(keywords):
    # Check what value the genre could be and convert it to the general general
    cleaned_keyword = keywords.lower().strip()
    if cleaned_keyword in country_map:
        country = country_map[cleaned_keyword] 
        return country
            
    return ""
```

Set the countries for the Map DF

```python
country_map_df = map_df.copy()
country_map_df['most_popular_country'] = country_map_df['most_popular_country'].apply(GetGeneralCountries)
display(country_map_df)
```

Set the coutnries for the dashboard age

```python
print(dashboard_Age_df["Country"])
```

```python
country_dashboard_age_df = dashboard_Age_df.copy()
country_dashboard_age_df['Country'] = country_dashboard_age_df['Country'].apply(GetGeneralCountries)
display(country_dashboard_age_df)
```

Get the dashboard Gender countries in
