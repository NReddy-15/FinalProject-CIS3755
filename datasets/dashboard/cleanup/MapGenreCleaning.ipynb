```python
# Grab the GetGeneralGenres function again 

def GetGeneralGenres(keywords):
    # Each user can have multiple favorite genres
    general_genres = []
    
    
    if  isinstance(keywords, str):
        # Get the current string of genres and split them into individual strings
        keywordsList = keywords.split(',')
        
        # Check what value the genre could be and convert it to the general general
        for word in keywordsList:
            if word in flat_map:
                genre = flat_map[word] 
                general_genres.append(genre)
                
        # Return a list of general genres
        return list(general_genres)
    return []
```

```python
# Grab a coopy of the map_df and then apply the general genres
combined_map_genre_df = map_df.copy()
combined_map_genre_df['genre'] = combined_map_genre_df['genre'].apply(GetGeneralGenres)
```

```python
# Check to ensure everything is ok
display(combined_map_genre_df)
```

```python
# Clean up the dataset
cleaned_combined_map_genre_df = combined_map_genre_df.copy()

cleaned_combined_map_genre_df = cleaned_combined_map_genre_df.explode("genre")
# cleaned_combined_map_genre_df["genre"] = cleaned_combined_map_genre_df["genre"].str.strip()

# cleaned_combined_map_genre_df["genre"] = cleaned_combined_map_genre_df["genre"].str.replace("[", "").str.replace("]", "")

display(cleaned_combined_map_genre_df)
```
