# For the Global Reads, match each user by their gender and country


```python
import pandas as pd
```

```python
# Get the dataframe 
# genreToUser_df = pd.read_csv("../old/GlobalReader/Cleaned_GlobalReader_GenreToUser.csv")
combined_df = pd.read_csv("../old/GlobalReader/Cleaned_GlobalReading_UserGenderToCountry_V2.csv")
```

```python
# display(genreToUser_df)
display(combined_df)
```

```python
# Combine both sets by User ID
# combined_df = pd.merge(genreToUser_df, genderCountryToUser_df, on="User_ID", how="inner")
```

```python
# display(combined_df)
```

```python
# Create a record for users who have 2+ genres in their genre list 
combined_explode_df = combined_df.copy()
combined_explode_df['Genres'] = combined_df["Genres"].str.split(',')
combined_explode_df = combined_explode_df.explode("Genres")
combined_explode_df["Genres"] = combined_explode_df["Genres"].str.strip()

combined_explode_df["Genres"] = combined_explode_df["Genres"].str.replace("'", "").str.replace("[", "").str.replace("]", "")

display(combined_explode_df)
```

```python
# Write out the final results
combined_explode_df.to_csv("../Combined_GlobalReader_GenreToGender_V2.csv")
```

# Creating a histogram to check the distribution

```python
# Check the distribution
temp_df = pd.read_csv("../Combined_GlobalReader_GenretoGender_V3.csv")
display(temp_df)
```

```python
temp_df["Country"].hist()
```

```python
# Create a matrix of Country vs Genre
table = pd.crosstab(temp_df['Country'], temp_df['Genres'])

print(table)
```

```python
# Filter by a specific genre and count records per country
genre_filter = 'Fantasy'  # Change this to any genre you want to filter by
genre_records = temp_df[temp_df['Genres'] == genre_filter]['Country'].value_counts()

print(f"Countries with records for '{genre_filter}':")
print(genre_records)
```
