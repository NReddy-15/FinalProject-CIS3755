```python
import pandas as pd
```

# Combine UserGender to their country of original

```python
# Load dataset
genderToCountry_df = pd.read_excel('../old/GlobalReading_UserGenderToCountry.xlsx')

# Display User_ID, Country, Gender
display(genderToCountry_df)
```

```python
# Remove any empty records from any column
genderToCountryCleaned_df = genderToCountry_df.dropna()

# 882528 users that do have a country
display(genderToCountryCleaned_df)
```

Create a new table that links users by their list of genres and their country + gender


```python
# Load the GenreToUser
genreToUser_df = pd.read_excel('../old/GlobalReader/Cleaned_GlobalReader_GenreToUser.xlsx')
display(genreToUser_df)
```

```python
# Merge the GenreToUser and GenderToCountry by User_ID
mergedGlobalReading = genreToUser_df.copy()
mergedGlobalReading = genderToCountryCleaned_df.merge(mergedGlobalReading[['User_ID', 'Genres']], on='User_ID', how='inner')
```

```python
# Displays User_ID, Country, Gender, Genres
display(mergedGlobalReading)
```

```python
# Remove any NaN
mergedGlobalReading.dropna()
display(mergedGlobalReading)
```

Export the Merged File
Will need one more layer of clean up to split the genres

```python
mergedGlobalReading.to_csv('../old/GlobalReader/Cleaned_GlobalReading_UserGenderToCountry_V2.csv', index=False)
```

## Checking the distribution of the data

```python
# Create a histogram to see the distribution of records

import matplotlib.pyplot as plt

mergedGlobalReading['Country'].value_counts().plot(kind='bar')

# Clean up the labels
plt.title('Number of Records per Country')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```

```python
# Then checked a minimum number of records by country. 
threshold = 30 # Iwant at least 30 records per country
country_counts = mergedGlobalReading['Country'].value_counts()
major_countries = country_counts[country_counts >= threshold].index

df_filtered = mergedGlobalReading[mergedGlobalReading['Country'].isin(major_countries)]
```
